### Implementation of A* and Comparison of A* with BFS

```python
class node:
    def __init__(self,parent,mat,x=-1,y=-1) -> None:
        self.misplaced = 0
        self.md = 0
        self.parent = parent
        self.mat = mat
        self.depth = 0
        self.myhash = hash((tuple(tuple(s) for s in self.mat)))
        self.lens = len(mat)
        if x == -1 and y == -1:
            for a in range(self.lens):
                for b in range(self.lens):
                    if self.mat[a][b] == -1:
                        self.x = a
                        self.y = b
        else:
            self.x = x
            self.y = y
    def CreateMyHash(self):
        self.myhash = hash((tuple(tuple(s) for s in self.mat)))
        return self
    def swap(self, a,b):
        temp = self.mat[self.x+a][self.y+b]
        self.mat[self.x+a][self.y+b] = self.mat[self.x][self.y]
        self.mat[self.x][self.y] = temp
        self.x+=a
        self.y+=b
        return self
    def printMat(self):
        for x in self.mat:
            print(x)
    def printMatLine(self):
        for x in self.mat:
            print(x,end=" ")
    def isEqual(self,target):
        if self.myhash != target.myhash:
            return False
        return True
    def CreateChild(self):
        child = node(self,[[x for x in y] for y in self.mat],self.x,self.y)
        child.depth = self.depth + 1
        return child
    def getParent(self):
        return self.parent
    def possibleLeft(self):
        if self.y > 0:
            return True
        else:
            return False
    def possibleRight(self):
        if self.y < self.lens-1:
            return True
        else:
            return False
    def possibleUP(self):
        if self.x > 0:
            return True
        else:
            return False
    def possibleDown(self):
        if self.x < self.lens-1:
            return True
        else:
            return False
    def goLeft(self):
        return self.swap(0,-1)
    def goRight(self):
        return self.swap(0,1)
    def goUP(self):
        return self.swap(-1,0)
    def goDown(self):
        return self.swap(1,0)
    def setMisplacement(self,val):
        self.misplaced = val
    def setMd(self,val):
        self.misplaced = val
    def __hash__(self):
        return self.myhash

class PQueue:
    def __init__(self):
        self.arr = []
    def append(self,n):
        if not self.arr:
            self.arr.append(n)
        else:
            for x in range(len(self.arr)):
                if self.arr[x].misplaced > n.misplaced:
                    self.arr.insert(x,n)
                    break
            self.arr.append(n)
    def pop(self)-> node:
        n = self.arr.pop(0)
        return n
    def inQueue(self,node):
        for x in self.arr:
            if x.isEqual(node):
                return True
        return False

class Stack:
    def __init__(self):
        self.arr = []
    def push(self,n):
        self.arr.append(n)
    def pop(self)-> node:
        n = self.arr.pop()
        return n
    def inStack(self,node):
        for x in self.arr:
            if x.isEqual(node):
                return True
        return False
    def Empty(self):
        return self.arr

class Visited:
    def __init__(self):
        self.arr = set()
    def visit(self,node):
        self.arr.add(node)
    def checkIfVisited(self,node):
        if node in self.arr:
            return True
        return False
    def printVisited(self):
        for x in self.arr:
            x.printMatLine()
            print()

class Queue:
    def __init__(self):
        self.arr = []
    def append(self,n):
        self.arr.append(n)
    def pop(self)-> node:
        n = self.arr.pop(0)
        return n
    def inQueue(self,node):
        for x in self.arr:
            if x.isEqual(node):
                return True
        return False
```

### Supporting Algo

```python

def getInvCount(arr,N):
    arr1=[]
    for y in arr:
        for x in y:
            arr1.append(x)
    arr=arr1
    inv_count = 0
    for i in range(N * N - 1):
        for j in range(i + 1,N * N):
            if (arr[j] and arr[i] and arr[i] > arr[j]):
                inv_count+=1


    return inv_count
def findXPosition(puzzle,N):
    for i in range(N - 1,-1,-1):
        for j in range(N - 1,-1,-1):
            if (puzzle[i][j] == -1):
                return N - i

def isSolvable(puzzle,N):
    invCount = getInvCount(puzzle,N)
    if (N & 1):
        return ~(invCount & 1)

    else:    # grid is even
        pos = findXPosition(puzzle,N)
        if (pos & 1):
            return ~(invCount & 1)
        else:
            return invCount & 1

def misPlacedTiles(initial,goal):
    mt = 0
    for x in range(len(initial.mat)):
        for y in range(len(initial.mat)):
            if initial.mat[x][y] != goal.mat[x][y]:
                mt += 1
    return mt

def manhattenDistance(grid_initial, grid_final):
    size = len(grid_initial.mat)
    total_distance = 0
    initial_positions = {}
    final_positions = {}
    for i in range(size):
        for j in range(size):
            initial_positions[grid_initial.mat[i][j]] = (i, j)
            final_positions[grid_final.mat[i][j]] = (i, j)
    for tile in range(1, size*size):
        initial_pos = initial_positions[tile]
        final_pos = final_positions[tile]
        distance = abs(initial_pos[0] - final_pos[0]) + abs(initial_pos[1] - final_pos[1])
        total_distance += distance
    return total_distance

def PrintQueue(queue):
    for x in range(len(queue.arr)%3):
        queue.arr[x].printMat()
```

### BFS

```python
def BFSAdvanced(initial,goal):
    nodesgen = 0
    if isSolvable(initial.mat,len(initial.mat)):
        pass
    else:
        print("Not solvable")
        return None
    queue = Queue()
    visited = Visited()
    queue.append(initial)
    while queue:
        if nodesgen > 100000:
            break
        state = queue.pop()
        visited.visit(state)
        #state.printMatLine()
        #print()
        if initial.isEqual(goal):
            return state,nodesgen
        else:
            if state.possibleRight():
                nodesgen+=1
                s1 = state.CreateChild().goRight().CreateMyHash()
                if s1.isEqual(goal):
                    return s1,nodesgen
                else:
                    if visited.checkIfVisited(s1) or queue.inQueue(s1):
                        pass
                    else:
                        queue.append(s1)
            if state.possibleLeft():
                nodesgen+=1
                s1 = state.CreateChild().goLeft().CreateMyHash()
                if s1.isEqual(goal):
                    return s1,nodesgen
                else:
                    if visited.checkIfVisited(s1) or queue.inQueue(s1):
                        pass
                    else:
                        queue.append(s1)
            if state.possibleUP():
                nodesgen+=1
                s1 = state.CreateChild().goUP().CreateMyHash()
                if s1.isEqual(goal):
                    return s1,nodesgen
                else:
                    if visited.checkIfVisited(s1) or queue.inQueue(s1):
                        pass
                    else:
                        queue.append(s1)
            if state.possibleDown():
                nodesgen+=1
                s1 = state.CreateChild().goDown().CreateMyHash()
                if s1.isEqual(goal):
                    return s1,nodesgen
                else:
                    if visited.checkIfVisited(s1) or queue.inQueue(s1):
                        pass
                    else:
                        queue.append(s1)
```
